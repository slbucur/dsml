# Categorical plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='sex', y='total_bill', data=tips) # average estimator as default
```

```python
import numpy as np
```

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std) # custom estimator (vec -> float)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='violin')
```

```python
sns.heatmap(fp, cmap='coolwarm', linecolor='white', linewidths=1)
```

```python
sns.clustermap(fp, cmap='coolwarm')
```

```python
sns.clustermap(fp, cmap='coolwarm', standard_scale=1)
```

# Grids

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
import matplotlib.pyplot as plt
g = sns.PairGrid(iris)
```

```python
# Then you map to the grid
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
# Map to upper,lower, and diagonal
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.distplot, 'total_bill')
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(plt.scatter, 'total_bill', 'tip')
```

# Regression plots

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'],
          scatter_kws={'s': 100})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex', aspect=0.6, size=8)
```
